```python
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt

gamma = 1.0
k = 48.

c1 = 1./6
c2 = 2./6
c3 = 2./6
c4 = 1./6
a2 = 1./2
a3 = 1./2
a4 = 1.
b21 = 1./2
b31 = 0.
b32 = 1./2
b41 = 0.
b42 = 0.
b43 = 1.


F_over_m = lambda t: 0
fxy_x = lambda t, x, v: v
fxy_v = lambda t, x, v: F_over_m(t)-2*gamma*v-k*x

def rk4_x(ti, xi, vi, dt):
    K1 = fxy_x(ti,xi, vi)
    K2 = fxy_x(ti+a2*dt,xi+b21*K1*dt,vi)
    K3 = fxy_x(ti+a3*dt,xi+b31*K1*dt+b32*K2*dt,vi)
    K4 = fxy_x(ti+a4*dt,xi+b41*K1*dt+b42*K2*dt+b43*K3*dt,vi) 
    xip1 = xi+(c1*K1+c2*K2+c3+K3*c4+K4)*dt
    return xip1

ti = 0.0
tf = 5.0
n = 5000
dt = (tf-ti)/n
t = np.zeros(n+1)
x = np.zeros(n+1)
v = np.zeros(n+1)
x_iv = 1.0
v_iv = -0.4
t[0] = ti
x[0] = x_iv
v[0] = v_iv

for i in range(1, n+1):
    t[i] = ti + i*dt
    x[i] = rk4_x(t[i-1],x[i-1],v[i-1],dt)
    v[i] = rk4_x(t[i-1],x[i-1],v[i-1],dt)

plt.plot(t,x,label="x vs t",color="blue",linewidth="3.0")
plt.title("Displacement for an underdamped system")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.tight_layout()
plt.show()    

plt.plot(t,v,label="v vs t",color="red",linewidth="3.0")
plt.title("Velocity for an underdamped system")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.tight_layout()
plt.show()
```

### Cosine Function

```python
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt

gamma = 10
k = 48.

c1 = 1./6
c2 = 2./6
c3 = 2./6
c4 = 1./6
a2 = 1./2
a3 = 1./2
a4 = 1.
b21 = 1./2
b31 = 0.
b32 = 1./2
b41 = 0.
b42 = 0.
b43 = 1.

F_over_m = lambda t: cos(2*pi*sqrt(k)*t)
fxy_x = lambda t, x, v: v
fxy_v = lambda t, x, v: F_over_m(t)-2*gamma*v - k*x

def rk4_x(ti, xi, vi, dt):
    K1 = fxy_x(ti, xi, vi)
    K2 = fxy_x(ti+a2*dt, ti+b21*K1*dt, vi)
    K3 = fxy_x(ti+a3*dt, xi+b31*K1*dt+b32*K2*dt, vi)
    K4 = fxy_x(ti+a4*dt,xi+b41*K1*dt+b42*K2*dt+b43*K3*dt, vi)
    xip1 = xi+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
    return xip1

a = 0.0 
b = 5.0
n = 5000
h = (b-a)/(n-1)
ti = 0.0
tf = 5.0
dt = (tf - ti)/n
t = np.zeros(n+1)
x = np.zeros(n+1)
v = np.zeros(n+1)
x_iv = 0.0
v_iv = 0.005
t[0] = a
x[0] = x_iv
v[0] = v_iv

for i in range(1, n+1):
    t[i] = a + i*dt
    x[i] = rk4_x(t[i-1], x[i-1], v[i-1], h)
    v[i] = rk4_v(t[i-1], x[i-1], v[i-1], h)

plt.plot(t,x,label="x vs t",color="blue",linewidth="3.0")
plt.title("Displacement for Cos system")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.tight_layout()
plt.show()    

plt.plot(t,v,label="v vs t",color="red",linewidth="3.0")
plt.title("Velocity for Cos system")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.tight_layout()
plt.show()
```
